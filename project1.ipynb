```python
import plotly.express as px
import matplotlib.pyplot as plt
df.groupby('brand')['stars'].sum().sort_values(ascending=False).head(10)
df.groupby('reviewsCount')['brand'].sum().sort_values(ascending=True).head(10)
df.reviewsCount<=10
amazon_kitchenware = pd.DataFrame(df)
```

```python
# Count the frequency of different methods
method_counts = amazon_kitchenware['brand'].value_counts()
```

```python
# Plotting the bar chart
plt.figure(figsize=(10, 6))
```

```python
plt.bar(method_counts.index, method_counts.values, color='violet')
plt.xlabel('brand')
plt.ylabel('stars')
plt.title('graph1')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
amazon_kitchenware= pd.DataFrame(df)

```

```python
# Plotting the pie chart
plt.figure(figsize=(8, 8))
amazon_kitchenware['stars'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightgreen',
'lightcoral', 'orange'])
plt.title(' pie chart')
plt.ylabel('')
plt.show()
amazon_kitchenware = pd.DataFrame(df)
```
